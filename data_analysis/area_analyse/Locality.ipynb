In this part I will try to analyse more in depth to effect of the locality on the Price

```python
import os
import warnings
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sn

file_path = r"C:\Users\Sacha\Documents\BeCode\real-estate-price-prediction\data_analysis\analyse_csv.csv" 
df_final = pd.read_csv(file_path)
```

```python
file_path = r"C:\Users\Sacha\Documents\BeCode\real-estate-price-prediction\data_analysis\area_analyse\square_meter_csv.csv" 
df_square_meter = pd.read_csv(file_path)
```

```python
freq_loc = df_final["Locality"].value_counts()

threshold = freq_loc.sum() * 0.005
freq_loc = freq_loc[freq_loc >= threshold]

plt.pie(freq_loc, labels=freq_loc.index)
plt.title("distribution of the location (zip code) in the data (>1%)")
plt.show()
```

In the previous graph we can see that some Locality have a big part of the data, we have to keep it in mind. After that we will plot the median prices by square meter for the localities, these prices was computed in the file price_by_square_m.ipynb  

```python
df_square_meter = df_square_meter.sort_values(by=['Median_Price_Squared_Meter'], ascending=False)
best_10 = df_square_meter.head(20).reset_index()
worst_10 = df_square_meter.tail(20).reset_index()
```

```python
plt.figure(figsize=(20, 8))
sn.barplot(x='Locality', y="Median_Price_Squared_Meter", data=best_10, order = best_10['Locality'].tolist())

plt.title("20 highest median square meter price")
plt.xlabel("Locality")
plt.ylabel("Price/square meter")

plt.savefig(r'.\images\high_price.jpg', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sn.barplot(x='Locality', y="Median_Price_Squared_Meter", data=worst_10, order = worst_10['Locality'].tolist())

plt.title("20 lowest median square meter price ")
plt.xlabel("Locality")
plt.ylabel("Price/square meter")

plt.savefig(r'.\images\low_price.jpg', bbox_inches='tight')

plt.show()
```
