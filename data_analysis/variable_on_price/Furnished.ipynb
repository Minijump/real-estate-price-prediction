# effect of the 'furnished' parameter on price

```python
import os
import warnings
import pandas as pd

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

col_names =  ["Locality",     
              "Type_of_property",
              "Subtype_of_property",
              "Price",
              "Type_of_sale",
              "Number_of_rooms",
              "Living_Area",
              "Fully_equipped_kitchen",
              "Furnished",
              "Open_fire",
              "Terrace",
              "Terrace_Area",
              "Garden",
              "Garden_Area",
              "Surface_area_of_the_plot_of_land",
              "Number_of_facades",
              "Swimming_pool",
              "State_of_the_building"
              ]

file_path = r"C:\Users\Sacha\Documents\BeCode\real-estate-price-prediction\data_analysis\main_csv.csv" 
df_final = pd.read_csv(file_path)
#df_final = df_final.dropna(subset=['Open_fire'])
```

```python
print(pd.unique(df_final["Furnished"]))
```

```python
perc =  df_final["Furnished"].isnull().sum() / len(df_final["Furnished"])
print(perc)
```

to see the 'impact' of nan, let replace it with a sting 'nan'

```python
df_final["Furnished"] = df_final["Furnished"].fillna("NAN")
```

```python
#if we consider nan = not installed:
#df_final = df_final.fillna("NOT_INSTALLED")
freq_type = df_final["Furnished"].value_counts()
plt.pie(freq_type, labels=freq_type.index, autopct='%1.1f%%')
plt.title("distribution of the Furnished in the data")
plt.show()
```

```python
df_nan = df_final[df_final['Furnished'] == "NAN"]
df_true = df_final[df_final['Furnished'] == True]
df_false = df_final[df_final['Furnished'] == False]

plt.figure(figsize=(20, 8))
plt.boxplot([df_nan['Price'], df_false['Price'], df_true['Price']], 
            showfliers=False, 
            labels=['NAN', 'False', 'True'])
plt.xlabel('kitchen', fontsize=15)
plt.ylabel("Price ()", fontsize=15)
plt.title("dist of price for house and appart (without outliers)")
plt.grid(True)

plt.show()
```
