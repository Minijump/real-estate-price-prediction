```python
import os
import warnings
import pandas as pd

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

col_names =  ["Locality",     
              "Type_of_property",
              "Subtype_of_property",
              "Price",
              "Type_of_sale",
              "Number_of_rooms",
              "Living_Area",
              "Fully_equipped_kitchen",
              "Furnished",
              "Open_fire",
              "Terrace",
              "Terrace_Area",
              "Garden",
              "Garden_Area",
              "Surface_area_of_the_plot_of_land",
              "Number_of_facades",
              "Swimming_pool",
              "State_of_the_building"
              ]

file_path = r"C:\Users\Sacha\Documents\BeCode\real-estate-price-prediction\data_analysis\main_csv.csv" 
df_final = pd.read_csv(file_path)
df_final = df_final.dropna(subset=['Surface_area_of_the_plot_of_land']) #at first drop the Nan, we will se after if we can make nan=0
```

```python
plt.boxplot(df_final['Surface_area_of_the_plot_of_land'])
plt.xlabel('Distribution', fontsize=15)
plt.ylabel("area (square_m)", fontsize=15)

plt.title('Distribution of Surface_area_of_the_plot_of_land', fontsize=18)
plt.grid(True)

plt.show()
```

```python
median_price_by_rooms = df_final.groupby("Surface_area_of_the_plot_of_land")["Price"].median().reset_index()
sns.scatterplot(x="Surface_area_of_the_plot_of_land", y="Price", data=median_price_by_rooms)
plt.show()
```

```python
# Compute the median price by number of rooms for each locality
groupby_area = df_final.groupby(["Locality", "Surface_area_of_the_plot_of_land"])["Price"].median().reset_index()

desired_localities = [1495, 4520]
groupby_area = groupby_area.loc[groupby_area['Locality'].isin(desired_localities)]

# Increase the size of the plot
plt.figure(figsize=(10,5))

# Create the scatter plot
sns.lmplot(x="Surface_area_of_the_plot_of_land", y="Price", hue="Locality", data=groupby_area, scatter_kws={'alpha':0.5}, aspect=1.5)

# Add x and y axis labels
plt.xlabel('Surface_area_of_the_plot_of_land (Square m)')
plt.ylabel('Median Price')

# Add a title to the plot
plt.title('Median Housing Prices by Locality and Surface_area_of_the_plot_of_land')

plt.show
```
