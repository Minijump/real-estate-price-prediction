```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
import seaborn.objects as so
```

```python
col_names =  ["Locality",     
              "Type_of_property",
              "Subtype_of_property",
              "Price",
              "Type_of_sale",
              "Number_of_rooms",
              "Living_Area",
              "Fully_equipped_kitchen",
              "Furnished",
              "Open_fire",
              "Terrace",
              "Terrace_Area",
              "Garden",
              "Garden_Area",
              "Surface_area_of_the_plot_of_land",
              "Number_of_facades",
              "Swimming_pool",
              "State_of_the_building"
              ]

file_path = r"C:\Users\Sacha\Documents\BeCode\real-estate-price-prediction\data_analysis\main_csv.csv" 
df_final = pd.read_csv(file_path)
df_final = df_final.dropna(subset=['Type_of_property'])
```

```python
def distr_type_of_property():
    """
    create a pie plot with the distribution of the different type of property
    """
    freq_type = df_final["Type_of_property"].value_counts()
    plt.pie(freq_type, labels=freq_type.index, autopct='%1.1f%%')
    plt.title("distribution of the type of property in the data")
    plt.show()
distr_type_of_property()
```

```python
is_house = df_final['Type_of_property'] == 'HOUSE'
df_house = df_final[is_house]
df_appart = df_final[~is_house]
```

```python
# Take a sample of size 400 from the 'Price' column
#len_sample = 500
#sample = df_house['Price'].sample(n = len_sample)

fig, ax = plt.subplots(1, 2, figsize=(18, 9))

ax[0].boxplot(df_house['Price'])
ax[0].set_xlabel('Distribution', fontsize=15)
ax[0].set_ylabel("Price (€)", fontsize=15)

#ax[0].set_title('Distribution of Price in a Sample of Size {} (houses)'.format(len_sample), fontsize=18)
ax[0].set_title('Distribution of Price (houses)', fontsize=18)
ax[0].grid(True)

ax[1].boxplot(df_house['Price'], showfliers=False)
ax[1].set_xlabel('Distribution', fontsize=15)
ax[1].set_ylabel("Price (€)", fontsize=15)

ax[1].set_title('Distribution of Price, Without Outliers (houses)', fontsize=18)
ax[1].grid(True)

plt.show()
```

```python
# Take a sample of size 400 from the 'Price' column

fig, ax = plt.subplots(1, 2, figsize=(18, 9))

ax[0].boxplot(df_appart['Price'])
ax[0].set_xlabel('Distribution', fontsize=15)
ax[0].set_ylabel("Price (€)", fontsize=15)

ax[0].set_title('Distribution of Price (appart)', fontsize=18)
ax[0].grid(True)

ax[1].boxplot(df_appart['Price'], showfliers=False)
ax[1].set_xlabel('Distribution', fontsize=15)
ax[1].set_ylabel("Price (€)", fontsize=15)

ax[1].set_title('Distribution of Price, Without Outliers (appart)', fontsize=18)
ax[1].grid(True)

plt.show()
```

```python
plt.boxplot([df_house['Price'], df_appart['Price']], showfliers=False)
plt.xlabel('House/appart', fontsize=15)
plt.ylabel("Price (€)", fontsize=15)
plt.title("dist of price for house and appart (without outliers)")
plt.grid(True)

plt.show()
```
