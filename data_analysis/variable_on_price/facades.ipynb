```python
import os
import warnings
import pandas as pd

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

col_names =  ["Locality",     
              "Type_of_property",
              "Subtype_of_property",
              "Price",
              "Type_of_sale",
              "Number_of_rooms",
              "Living_Area",
              "Fully_equipped_kitchen",
              "Furnished",
              "Open_fire",
              "Terrace",
              "Terrace_Area",
              "Garden",
              "Garden_Area",
              "Surface_area_of_the_plot_of_land",
              "Number_of_facades",
              "Swimming_pool",
              "State_of_the_building"
              ]

file_path = r"C:\Users\Sacha\Documents\BeCode\real-estate-price-prediction\data_analysis\main_csv.csv" 
df_final = pd.read_csv(file_path)
df_final = df_final.dropna(subset=['Number_of_facades']) #at first drop the Nan, we will se after if we can make nan=0
```

```python
freq_type = df_final["Number_of_facades"].value_counts()
plt.pie(freq_type, labels=freq_type.index, autopct='%1.1f%%')
plt.title("distribution of the Number_of_facades in the data")
plt.show()
```

```python
median_price_by_facades = df_final.groupby("Number_of_facades")["Price"].median().reset_index()

# Create the scatter plot
sns.scatterplot(x="Number_of_facades", y="Price", data=median_price_by_facades)
plt.show()
```
