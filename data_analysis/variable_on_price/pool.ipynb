```python
import os
import warnings
import pandas as pd

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

col_names =  ["Locality",     
              "Type_of_property",
              "Subtype_of_property",
              "Price",
              "Type_of_sale",
              "Number_of_rooms",
              "Living_Area",
              "Fully_equipped_kitchen",
              "Furnished",
              "Open_fire",
              "Terrace",
              "Terrace_Area",
              "Garden",
              "Garden_Area",
              "Surface_area_of_the_plot_of_land",
              "Number_of_facades",
              "Swimming_pool",
              "State_of_the_building"
              ]

file_path = r"C:\Users\Sacha\Documents\BeCode\real-estate-price-prediction\data_analysis\main_csv.csv" 
df_final = pd.read_csv(file_path)
```

```python
print(pd.unique(df_final["Swimming_pool"]))
```

```python
df_final["Swimming_pool"] = df_final["Swimming_pool"].fillna("NAN")
```

```python
freq_type = df_final["Swimming_pool"].value_counts()
plt.pie(freq_type, labels=freq_type.index, autopct='%1.1f%%')
plt.title("distribution of the Swimming_pool in the data")
plt.show()
```

```python
df_true = df_final[df_final["Swimming_pool"] == True]
df_nan = df_final[df_final["Swimming_pool"] == "NAN"]
df_false = df_final[df_final["Swimming_pool"] == False]

plt.figure(figsize=(20, 8))
plt.boxplot([df_nan['Price'], df_false['Price'], df_true['Price']], 
            showfliers=False, 
            labels=['NAN', 'False', 'True'])
plt.xlabel('oswiming pool', fontsize=15)
plt.ylabel("Price (€)", fontsize=15)
plt.title("dist of price for Swimming_pool (without outliers)")
plt.grid(True)

plt.show()
```
